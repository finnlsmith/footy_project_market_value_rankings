## Adding the lookup required matches

```python
# Assuming df is your DataFrame
def parse_season(input_year_test):
    if '.' in input_year_test:
        yearstr = input_year_test.split(".")[2]
        full_num = int('20' + yearstr)
    elif '-' in input_year_test:
        if len(input_year_test.split("-")[2]) == 4:
            yearstr = input_year_test.split("-")[2] #this is a 4 digit number
            full_num = int(yearstr)
        elif len(input_year_test.split("-")[2]) == 2:
            if len(input_year_test.split("-")[0]) == 4:
                yearstr = input_year_test.split("-")[0]
                full_num = int(yearstr)
            else:
                yearstr = input_year_test.split("-")[2] #this is a 2 digit number
                full_num = int('20' + yearstr)
        else:
            print('broke on year parsing', input_year_test.split("-"))
            full_num = None
    else:
        print('Unable to parse input:', input_year_test)
        full_num = None
    return full_num
```

```python
big_df_latest = load_csv_dataset('combined_df_merged_APRIL_22.csv')
big_df_latest = big_df_latest.drop(columns={'Unnamed: 0'})
```

```python
big_df_latest
```

```python
lookup_required_matches = load_csv_dataset('Group 1 and 2 (for T) - Find their $ Info/Terrell Lookup Cases - lookup_required_matches_success_rework.csv')
lookup_required_matches = lookup_required_matches.drop(columns={'Unnamed: 0.1', 'Unnamed: 0'})
lookup_required_matches.rename(columns={'Country Code': 'Team Country Code', 'Jersey Name': 'ORIGINAL JERSEY', 'Name Match': 'Name(s) Found'}, inplace=True)


lookup_required_matches['Match Case'] = 'single'
lookup_required_matches['Impute Required?'] = False

lookup_required_matches['Season'] = lookup_required_matches['Date'].apply(parse_season)

lookup_required_matches['Name'] = lookup_required_matches['Name(s) Found']
# Define the desired column order
desired_columns = ['Match ID', 'Competition', 'Match', 'Date', 'Nationality', 'Team Country Code', 
                   'Season', 'Status', 'Name', 'Market Value', 'Lookup Still Required?', 
                   'Lookup Return Case', 'Impute Required?', 'Name(s) Found', 'ORIGINAL JERSEY', 'Match Case']

# Reorder the columns
lookup_required_matches = lookup_required_matches[desired_columns]
lookup_required_matches
```

```python
combined_df = pd.concat([big_df_latest, lookup_required_matches], ignore_index=True)
```
