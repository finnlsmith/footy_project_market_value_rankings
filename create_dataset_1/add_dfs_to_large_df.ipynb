## Group 1 Combined

```python
big_df_latest = load_csv_dataset('most_updated_combined_DF.csv')
big_df_latest = big_df_latest.drop(columns={'Unnamed: 0'})
```

```python
big_df_latest
```

```python
group_1_working = load_csv_dataset('Group 1 and 2 (for T) - Find their $ Info/Terrell Lookup Cases - group1_success_rework.csv')
group_1_working = group_1_working.drop(columns={'Unnamed: 0.1', 'Unnamed: 0'})
group_1_working.rename(columns={'Names_Found': 'Name(s) Found'}, inplace=True)
group_1_working.rename(columns={'Country Code': 'Team Country Code'}, inplace=True)
group_1_working = group_1_working.assign(Name=group_1_working['Name(s) Found'])
group_1_working['Impute Required?'] = False
group_1_working['Match Case'] = 'single'

for i in range(0, 5260):
    if group_1_working.at[i, 'Market Value'] == 0:
        group_1_working.at[i, 'Impute Required?'] = True


desired_column_order = ['Match ID', 'Competition', 'Match', 'Date', 'Nationality', 'Team Country Code', 
                       'Season', 'Status', 'Name', 'Market Value', 'Lookup Still Required?', 
                       'Lookup Return Case', 'Impute Required?', 'Name(s) Found', 'ORIGINAL JERSEY', 
                       'Match Case']

group_1_working = group_1_working.reindex(columns=desired_column_order)
group_1_working
```

```python
big_df_merged_new_Df = pd.concat([big_df_latest, group_1_working], ignore_index=True)
big_df_merged_new_Df
```

```python
big_df_merged_new_Df.to_csv('big_df_merged(group 1 added).csv')
```
