```python
import pandas as pd
import os
import re

def load_csv_dataset(file_path):
    return pd.read_csv(file_path)
```

```python
os.getcwd()
```

```python
match_dataset_latest = load_csv_dataset('footy_project_market_value_rankings/create_dataset_1/CSVs we use often/edited_match_dataset_new_may_30.csv')
match_dataset_latest = match_dataset_latest.drop(columns={'...1', 'Unnamed..0'})
def get_substring_before_number(s):
    match = re.match(r'^[^\d]*', s)
    if match:
        return match.group(0)
    return s
def get_numbers(s):
    numbers = re.findall(r'\d+', s)
    return ''.join(numbers)
for index, row in match_dataset_latest.iterrows():
    if 'Olympics' in row['Match.ID']:
        competition = get_substring_before_number(row['Match.ID']) 
        nums = get_numbers(row['Match.ID'])
        match_dataset_latest.at[index, 'Competition'] = competition
        match_dataset_latest.at[index, 'Match.ID'] = f"{competition}_{nums}"
        
```
