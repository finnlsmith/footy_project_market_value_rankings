## Handle U-20 matches

```python
#LOOP THROUGH THE DATAFRAME RESULTS 
#IF THE WORD U-20 IS IN THE column called COMPETITION OF THAT ROW 
#ADD U20 TO THE END OF THE STRING FOR COLUMN "TEAM 1" AND FOR COLUMN "TEAM 2"

for i in range(0, len(results)):
    # Check if 'U-20' is present in the 'COMPETITION' column
    if(pd.isna(results.at[i,'Competition'])):
        0==0
    else:
        if 'U-20' in results.at[i,'Competition']:
            # Add 'U20' to the end of 'TEAM 1' and 'TEAM 2'
            results.at[i, 'Team 1'] += ' U-20'
            results.at[i, 'Team 2'] += ' U-20'

            #bit to change
            teams_names = results.at[i, 'SearchString'].split(' on ')[0]

            team_name_1 = teams_names.split(' vs ')[0]
            team_name_2 = teams_names.split(' vs ')[1]

            team_name_1 = team_name_1 + " U-20"
            team_name_2 = team_name_2 + " U-20"

            teams_names = team_name_1 + " vs " + team_name_2

            #bit to keep the same
            rest_of_searchstring = results.at[i, 'SearchString'].split(' on ')[1]

            #putting them together 
            results.at[i, 'SearchString'] = teams_names + " on "+ rest_of_searchstring 


```

## Fixing date columns

### Lineups

```python
to_drop_indexes = []
for index, row in lineups.iterrows():
    if(pd.isna(lineups.loc[index]['Starter_1'])):
        #lineups = lineups.drop(index, axis=0).reset_index().drop(columns={'index'})
        to_drop_indexes.append(index)
    else:
        date_str = lineups.loc[index]['Date']#was row['Date']
        try:
            # Try parsing the date using the first format
            date_obj = datetime.strptime(date_str, '%d.%m.%y')
        except ValueError:
            try:
                # Try parsing the date using the second format
                date_obj = datetime.strptime(date_str, '%d-%b-%y')
                # Convert the date to the required format
                lineups.at[index, 'Date'] = date_obj.strftime('%d.%m.%y')
            except ValueError:
                try:
                    # Try parsing the date using the third format
                    date_obj = datetime.strptime(date_str, '%d-%b-%y   %H:%M')
                    # Convert the date to the required format without the time part
                    lineups.at[index, 'Date'] = date_obj.strftime('%d.%m.%y')
                except ValueError:
                    # If parsing fails for all formats, leave the date as is
                    pass

to_drop_indexes.append(2089)
```

```python

existing_indexes = set(lineups.index)
to_drop = [index for index in to_drop_indexes if index in existing_indexes]

if to_drop:
    # Drop the specified indexes
    lineups.drop(to_drop, inplace=True)

lineups = lineups.reset_index().drop(columns={'index'})
```
