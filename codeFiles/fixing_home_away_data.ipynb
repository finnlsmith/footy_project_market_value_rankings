```python
def compare_teams(match, team_1_name, team_2_name):
    team_1, team_2 = match.split(' vs ')
    
    team_1_match = team_1.strip() == team_1_name.strip()
    team_2_match = team_2.strip() == team_2_name.strip()
    
    if team_1_match and team_2_match:
        return 2, match
    elif team_1_match or team_2_match:
        # Update team_1 in the match string if it doesn't match team_1_name
        if team_1.strip() != team_1_name.strip():
            team_1 = team_1_name.strip()
        
        # Update team_2 in the match string if it doesn't match team_2_name
        if team_2.strip() != team_2_name.strip():
            team_2 = team_2_name.strip()
        
        # Construct the new match string
        new_match = f"{team_1} vs {team_2}"
        return 1, new_match
    else:
        #if team_1.strip() != team_1_name.strip() and team_2.strip() != team_2_name.strip():
        return 0, match


# Example usage for a single row
# row = latest_match_data.iloc[0]
# match = row['Match']
# team_1_name = row['Team.1.Name']
# team_2_name = row['Team.2.Name']


for index, row in latest_match_data.iterrows():
    match = row['Match']
    team_1_name = row['Team.1.Name']
    team_2_name = row['Team.2.Name']
    result, updated_match = compare_teams(match, team_1_name, team_2_name)
    latest_match_data.at[index, 'Home_Away_correct_match'] = result
    
    # Update the 'Match' column with the new match string if it has changed
    if updated_match != match:
        latest_match_data.at[index, 'Match'] = updated_match
# print(f"Result: {result}")
```

```python
columns_to_swap = [
    ('Team.1.Players', 'Team.2.Players'),
    ('Team.1.Players_Positions', 'Team.2.Players_Positions'),
    ('Team.1.Code', 'Team.2.Code'),
    ('Team.1.Name', 'Team.2.Name'),
    ('Team.1.Values.By.Player', 'Team.2.Values.By.Player'),
    ('Team.1.Confederation', 'Team.2.Confederation'),
    ('Team.1.Score', 'Team.2.Score'),
    ('Team.1.Rank', 'Team.2.Rank'),
    ('Team.1.Total.Market.Value', 'Team.2.Total.Market.Value'),
    ('Team.1.Num.of.0.Value.Players', 'Team.2.Num.of.0.Value.Players'),
    ('Team 1, Tier 1 Global Superstars', 'Team 2, Tier 1 Global Superstars'),
    ('Team 1, Tier 2 Global Superstars', 'Team 2, Tier 2 Global Superstars'),
    ('Team 1, Tier 1 Confederation Stars', 'Team 2, Tier 1 Confederation Stars'),
    ('Team 1, Tier 2 Confederation Stars', 'Team 2, Tier 2 Confederation Stars'),
    ('Team 1, Total Global Superstars', 'Team 2, Total Global Superstars'),
    ('Team 1, Total Confederation Stars', 'Team 2, Total Confederation Stars')
]

# Define the columns to multiply by -1
columns_to_multiply = [
    'Rank.Difference',
    'Market.Value.Difference',
    'Number.of.Valuable.Players.Difference',
    'Standardized_Rank_Difference',
    'Global_Superstar_difference',
    'Confederation_star_difference'
]

# Define the columns to swap values
columns_to_swap_values = [
    ('Number_of_valuable_players_prediction_correct', 'Number_of_valuable_players_prediction_correct_inverse'),
    ('MV_Prediction_Correct', 'MV_Prediction_Correct_inverse'),
    ('Global_Superstar_difference_prediction_correct', 'Global_Superstar_difference_prediction_correct_inverse'),
    ('rank_prediction_correct', 'Rank_Prediction_Correct_inverse'),
    ('Team.1.Wins', 'Team.2.Wins')
]

# Iterate over rows where Home_Away_correct_match is 0
for index, row in latest_match_data[latest_match_data['Home_Away_correct_match'] == 0].iterrows():
    # Swap the columns
    for col1, col2 in columns_to_swap:
        latest_match_data.at[index, col1], latest_match_data.at[index, col2] = row[col2], row[col1]
    
    # Multiply the columns by -1
    for col in columns_to_multiply:
        latest_match_data.at[index, col] = row[col] * -1
    
    # Swap the values in the specified columns
    for col1, col2 in columns_to_swap_values:
        latest_match_data.at[index, col1], latest_match_data.at[index, col2] = row[col2], row[col1]
    
    # # Swap the score
    # score = row['Team.1.Score'] + ':' + row['Team.2.Score']
    # new_score = f"{score.split(':')[1]}:{score.split(':')[0]}"
    # latest_match_data.at[index, 'Team.1.Score'], latest_match_data.at[index, 'Team.2.Score'] = new_score.split(':')
```
